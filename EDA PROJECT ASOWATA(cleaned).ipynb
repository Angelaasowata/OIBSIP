```python


import pandas as pd
data = pd.read_csv("retail_sales_dataset.csv")
```

```python
data.groupby('Product Category')['Total Amount'].sum()
```

```python
data.groupby('Gender')['Total Amount'].sum()
```

```python
data.groupby('Gender')['Total Amount'].sum().plot(kind='bar')
plt.title('Total Sales by Gender')
plt.show()
```

```python
data['Age'].plot(kind='hist')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
data.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Month'] = data['Date'].dt.month_name()

data.groupby('Month')['Total Amount'].sum().plot(kind='bar')
plt.title('Monthly Sales')
plt.show()
```
